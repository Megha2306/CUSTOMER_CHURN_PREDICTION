```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv(r"C:\Users\ADMIN\Downloads\Cus. Churn Prediction-20240511T130240Z-001\Cus. Churn Prediction\Churn_Modelling.csv")
```

```python
data.head()
```

```python
col=['CustomerId','Surname','RowNumber']
d1=data.drop(columns=col)
d1.head()
```

```python
d1.info()
```

```python
categorical_columns = d1.select_dtypes(include=['object']).columns
d2=d1.copy()
label_encoder = LabelEncoder()

for cols in categorical_columns:
    d2[cols] = label_encoder.fit_transform(d2[cols])
    
```

```python
d2.head()
```

```python
d3=d2.copy()
corr_matrix = d3.corr()
corr_matrix
```

```python
high_corr_vars = set()
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > 0.6:
            colname = corr_matrix.columns[i]
            high_corr_vars.add(colname)
```

```python
d3= d3.drop(columns=high_corr_vars)
d3.info()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['variable']= d3.columns

vif['vif'] = [variance_inflation_factor(d3.values,i)
 for i in range(d3.shape[1])]

vif
```

```python
x = d3.iloc[:,:-1].values
y = d3.iloc[:,-1].values

x=pd.DataFrame(x)
dependant_col='Churn'
col=[cols for cols in d3.columns if cols != dependant_col]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,
 test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
reg_log = LogisticRegression(solver='lbfgs', max_iter=5000)
reg_log.fit(x_train,y_train)
```

```python
y_pred_log = reg_log.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

# Calculate accuracy using accuracy_score
accuracy = accuracy_score(y_test, y_pred_log)
print(f"Accuracy: {accuracy}")
```

```python
from sklearn.tree import DecisionTreeClassifier

clf_entropy = DecisionTreeClassifier(criterion="entropy", random_state=100,max_depth=3,min_samples_leaf=5)
 
    
clf_entropy.fit(x_train, y_train)
y_pred_decision_tree = clf_entropy.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred_decision_tree)
accuracy
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred_random_forest = rf.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred_random_forest)
accuracy
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
gb_model.fit(x_train, y_train)
```

```python
y_pred_gb = gb_model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred_gb)
print(f"Accuracy: {accuracy}")
```
